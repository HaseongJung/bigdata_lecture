```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.family'] ='Malgun Gothic'    # 한글 깨짐 방지
plt.rcParams['axes.unicode_minus'] =False  
```

```python
# data
x = [23.1, 32.8, 31.8, 32.0, 30.4, 24.0, 37.9, 30.5, 25.1, 12.4, 35.1, 31.5, 39.5, 24.2, 52.5, 21.1, 27.6]
y = [10.5, 17.0, 18.2, 17.0, 16.3, 10.5, 22.8, 14.1, 12.9, 8.8, 17.4, 14.9, 23.1, 12.4, 24.9, 10.5, 16.1]

data = pd.DataFrame({
    'x':x,
    'y':y
    })

data
```

##### (a). 선형적인 연관관계가 있는지를 산점도를 통해서 확인하여라.

```python
# sactter plot

plt.scatter(data.x, data.y)
plt.xlabel("수분 함유량")
plt.ylabel("용수량")
plt.show()
```

```python
# joint plot
import seaborn as sns

sns.jointplot(x=x, y=y, kind='reg', data=data)
plt.xlabel("수분 함유량")
plt.ylabel("용수량")
```
